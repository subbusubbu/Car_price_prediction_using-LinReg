```python
data = pd.read_csv("/car_price_dataset.csv")
```

```python
#Feature Selection
#selecting Independent and Dependent Features
x = pd.DataFrame(data.iloc[:,:-1])
y = pd.DataFrame(data.iloc[:,-1])
```

```python
categorical_columns = ["Brand", "Model", "Fuel_Type", "Transmission"]
```

```python
data = pd.get_dummies(data, columns= categorical_columns, drop_first=True)
```

```python
x = data.drop(columns=["Price"])
y = data["Price"]
```

```python
x.dtypes
```

```python
from matplotlib import pyplot as plt
```

```python
#Diving data into training and test data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
#printing the shape of training data and splitting data
print(f"The shape of x_train : f{x_train.shape} ")
print(f"The shape of y_train : f{y_train.shape}")
print("-"*30)
print(f"The shape of x_test : f{x_test.shape}")
print(f"The shape of y_test : f{y_test.shape}")
```

```python
#model building
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model
```

```python
#fitting in to model
model.fit(x_train,y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
#Now start predictions
y_pred = model.predict(x_test)
y_pred = pd.DataFrame(y_pred, columns=["Predicted"])
```

```python
y_pred
```

```python
y_test
```

```python
#finding Error
from sklearn import metrics
print(f"MAE : {metrics.mean_absolute_error(y_test,y_pred)}")
print(f"MSE : {metrics.mean_squared_error(y_test,y_pred)}")
print(f"RMSE: {np.sqrt(metrics.mean_squared_error(y_test,y_pred))}")
```

```python
#finding the r2(which is coefficient of determination)
print(f"r2 is : {metrics.r2_score(y_test,y_pred)}")
```

**We got 99% accuracy in finding the price of the car using historical data**
